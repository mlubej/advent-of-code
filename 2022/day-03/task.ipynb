```python
import re

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from aocd import get_data, submit

DAY = 3
YEAR = 2022

data = get_data(day=DAY, year=YEAR)
# print(data)
```

```python
def parse_data(data):
    data = data.split("\n")
    return data

data = parse_data(data)
```

# Part 1

```python
def item_score(ch):
    if ch.islower():
        return ord(ch) - 96
    else:
        return ord(ch) - (64 - 26)


def calc_score(line):
    l, r = line[: int(len(line) / 2)], line[int(len(line) / 2) :]
    for ch in l:
        if ch in r:
            return item_score(ch)


# calc_score('vJrwpWtwJgWrhcsFMMfFFhFp')
result = sum([calc_score(d) for d in data])
result
```
